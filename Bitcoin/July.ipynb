```python
#Importando a função criada no arquivo #pegar_historico.py
from pegar_historico import pegar_historico
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python


#Testando a função para o par de trding BTCBRL
par = "BTCBRL"
intervalo = "1d"
data_inicio = "1 Jan 2025"

df_btcbrl = pegar_historico(
        par,
        intervalo,
        data_inicio
    )
```

```python
valores_julhos = df_btcbrl[
    df_btcbrl['timestamp'] >= '2025-07-01'
]
```

```python
valores_junho = df_btcbrl[
    (df_btcbrl['timestamp'] >= '2025-06-01') & (df_btcbrl['timestamp'] <= '2025-06-30')
]

valores_junho
```

```python
px.line(
    valores_julhos,
    x='timestamp',
    y='close',
    width=1200,    
    height=600    
)
```

```python
#comparando com o gráfico do mês passado
fig = px.line(
    valores_junho,
    x='timestamp',
    y='close',
    width=1200,   
    height=600
    
)

fig.update_traces(line=dict(color='orange'))
```

```python
#comparando com o gráfico do mês passado
fig = px.line(
    df_btcbrl,
    x='timestamp',
    y='close',
    width=1200,   
    height=600
    
)

fig.update_traces(line=dict(color='orange'))
```

```python
df_btcbrl['Mês'] = df_btcbrl['timestamp'].dt.strftime('%m')
```

```python
data = pd.DataFrame(
    df_btcbrl.groupby('Mês')['close'].mean(),
    
)

import numpy as np

data = data.reset_index()
np.round(data, 3)
```

```python
data['Mês'] = data['Mês'].map({
    
    '01':'Janeiro',
    '02':'Fevereiro',
    '03':'Março',
    '04':'Abril',
    '05':'Maio',
    '06':'Junho',
    '07':'Julho'
    }
)

data
```

```python
plt.figure(
    figsize=(8,5)
)

sns.barplot(
    data=data,
    x='Mês',
    y='close',
    palette='Set2'
)

plt.title('Preço médio do bitcoint por mês em 2025')
```
