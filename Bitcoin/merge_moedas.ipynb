```python
import pandas as pd
from pegar_historico import pegar_historico
```

```python
#Pegando os valores de btc brl
par = "BTCBRL"
intervalo = "1d"
data_inicio = "1 Jun 2025"

df_btcbrl = pegar_historico(
        par,
        intervalo,
        data_inicio
    )

cols = ['timestamp', 'open', 'high', 'low', 'close']

ticker  = []

for i in range(0, len(df_btcbrl)):
    ticker.append('BTC')
    
df_btcbrl = df_btcbrl[cols]
df_btcbrl['coin'] = ticker
df_btcbrl.head(5)
```

```python
#Pegando os valores de btc brl
par = "SOLBRL"
intervalo = "1d"
data_inicio = "1 Jun 2025"

df_solbrl = pegar_historico(
        par,
        intervalo,
        data_inicio
    )

cols = ['timestamp', 'open', 'high', 'low', 'close']

ticker  = []

for i in range(0, len(df_solbrl)):
    ticker.append('SOLANA')
    
df_solbrl = df_solbrl[cols]
df_solbrl['coin'] = ticker
df_solbrl.head(5)
```

```python
coinsdf = pd.concat([
    df_btcbrl,
    df_solbrl]
)
```

```python
coinsdf.groupby('coin').max()
```
