```python
import matplotlib.pyplot as plt
import plotly.express as px
from binance.client import Client
import seaborn as sns
import numpy as np
from pegar_historico import pegar_historico
```

```python
par = "BTCBRL"
intervalo = "1d"
data_inicio = "1 Jan 2025"

df_btcbrl = pegar_historico(
        par,
        intervalo,
        data_inicio
    )

df_btcbrl.head(5)
```

```python
# Seleciona apenas as colunas 'close' e 'high' do DataFrame original (df_btcbrl)
# Essas colunas representam o pre√ßo de fechamento e o pre√ßo m√°ximo de cada candle (geralmente di√°rio ou por minuto)
df_ch = df_btcbrl[['close', 'high']]

# Cria uma nova coluna chamada 'diferenca' no DataFrame df_ch
# Essa coluna calcula a diferen√ßa percentual entre o valor m√°ximo (high) e o valor de fechamento (close)
# A f√≥rmula √©: ((high - close) / high) * 100
# Isso indica o quanto o pre√ßo de fechamento ficou abaixo do topo do candle, em porcentagem
df_ch['diferenca'] = (
    (df_ch['high'] - df_ch['close']) / df_ch['high']
) * 100

# Exibe as 10 primeiras linhas do DataFrame df_ch, j√° com a nova coluna 'diferenca'
# √ötil para verificar os valores calculados e garantir que tudo est√° correto
df_ch.head(10)
```

```python
#üß† O que est√° acontecendo aqui:
# Acessa a coluna 'diferenca' que voc√™ criou anteriormente, contendo a diferen√ßa percentual entre o pre√ßo m√°ximo (high) e o pre√ßo de fechamento (close) 
#de cada candle.

#.mean() Calcula a m√©dia (m√©dia aritm√©tica) de todos os valores dessa coluna. Ou seja, est√° respondendo √† pergunta:
#‚ÄúEm m√©dia, quanto o pre√ßo de fechamento ficou abaixo do topo dos candles analisados?‚Äù

round(
    df_ch['diferenca'].mean(),
    4
)
```
