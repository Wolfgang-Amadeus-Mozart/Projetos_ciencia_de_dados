```python
import matplotlib.pyplot as plt
import plotly.express as px
from binance.client import Client
import seaborn as sns
import numpy as np
from pegar_historico import pegar_historico
```

```python
par = "BTCBRL"
intervalo = "1d"
data_inicio = "1 Jan 2025"

df_btcbrl = pegar_historico(
        par,
        intervalo,
        data_inicio
    )

df_btcbrl.head(5)
```

```python
# Seleciona apenas as colunas 'close' e 'high' do DataFrame original (df_btcbrl)
# Essas colunas representam o preço de fechamento e o preço máximo de cada candle (geralmente diário ou por minuto)
df_ch = df_btcbrl[['close', 'high']]

# Cria uma nova coluna chamada 'diferenca' no DataFrame df_ch
# Essa coluna calcula a diferença percentual entre o valor máximo (high) e o valor de fechamento (close)
# A fórmula é: ((high - close) / high) * 100
# Isso indica o quanto o preço de fechamento ficou abaixo do topo do candle, em porcentagem
df_ch['diferenca'] = (
    (df_ch['high'] - df_ch['close']) / df_ch['high']
) * 100

# Exibe as 10 primeiras linhas do DataFrame df_ch, já com a nova coluna 'diferenca'
# Útil para verificar os valores calculados e garantir que tudo está correto
df_ch.head(10)
```
