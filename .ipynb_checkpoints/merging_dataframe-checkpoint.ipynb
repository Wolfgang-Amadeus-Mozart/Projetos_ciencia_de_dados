## Juntando daframe de tamanhos iguais

```python
#Importações

import pandas as pd
```

```python
#Criando dataframe 1
df1 = pd.DataFrame(
            {
                'A':range(1,11)
            }
        )

#Criando dataframe2
df2 =  pd.DataFrame(
            {
                'A':range(1,11),
                'B':range(11,21)
            }
        )

#juntando os dataframe
pd.merge(df1, df2)
```

## juntando daframe de tamanhos diferentes

```python
#Criando dataframe 1
df1 = pd.DataFrame(
            {
                'A':range(1,12) #Defining 1 valor  adcional para o dataframe 1
            }
        )

#Criando dataframe2
df2 =  pd.DataFrame(
            {
                'A':range(1,11),
                'B':range(11,21)
            }
        )

#juntando os dataframe
pd.merge(df1, df2)
```

Percebe-se que o valor adcional que foi colocado no dataframe 1 nao foi adcionando ao merge, porem, podemos definir o lado da junção

```python
pd.merge(df1, 
         df2,
        how='left' #Defining um merge a direita
        )
```

Agora o valor do df1 foi adcionado, e um valor NaN foi adcionado a coluna da direita para compensar o valor faltante

```python
#Criando dataframe 1
df1 = pd.DataFrame(
            {
                'A':range(1,12) #Defining 1 valor  adcional para o dataframe 1
            }
        )

#Criando dataframe2
df2 =  pd.DataFrame(
            {
                'A':range(1,11),
                'B':range(11,21)
            }
        )

df3 = pd.DataFrame(
            {
                'A':range(8,12),
                'B':range(15,19),
                'C':range(35,39)
            }
        )

#Como unir 3 dataframe
merge = pd.merge(df1, 
             pd.merge(df2, df3, how='left', on='A'),
             how='left'
            )

merge
```

```python
#Seleção com merge

merge[ 
~ merge.isin( # ~ significa 'not', neste caso estamos procurando o valor que nao esta no merge entre df2 e df3
    pd.merge(
        df2, 
        df3, 
        how='left', 
        on='A')
)]
```

```python
#Rodando o merge de df2 e df3 para comprovar visualmente que de fato a valor '11' é o unico valor que nao aparece entre esse merge
pd.merge(
        df2, 
        df3, 
        how='left', 
        on='A')
```
