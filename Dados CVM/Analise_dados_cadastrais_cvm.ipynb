```python
#importações 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import chardet
```

```python
#Foi necessario utilizar a biblioteca chardet para detectar o tipo de encoding do arquivo, pois a biblioteca pandas estava falha em ler o arquivo

with open("datasets/CVM Dados Cadastrais.csv", "rb") as file:
    result = chardet.detect(file.read())
    print(result)  # Shows the detected encoding
    encoding = result['encoding']
```

```python
df = pd.read_csv('datasets/CVM Dados Cadastrais.csv', sep=';', encoding=encoding)
df
```

```python
#Formato do dataframe e informações básicas
print(f'Numero de observações : {df.shape[0]} \nNumero de variaveis {df.shape[1]}')
df.info()
```

```python
#Após verificar as primeiras impressões, constatou-se uma grande quantidade de dados nulos, mas será necessario avaliar quanto exatamenteD
#Aqui um dicionario que cotem todas as variaveis do dataframe como chave e como valor contem a contagem de dados nulos

dici = {}

for i in df.columns:
    dici[i] = len(df[i][df[i].isna() == True])

dici
```

```python
df_isna = pd.DataFrame(dici.values(), index = dici.keys(), columns=['Valores_Faltantes']) #Cria um dataframe utilizando as chaves do dicionario como index
#E os valores do dicionario como obersações

df_isna['Valores_faltantes_relativos'] = np.round((df_isna['Valores_Faltantes'] / df.shape[0]) * 100, decimals = 2) #Cria uma nova variavel
#Os valores dessa nova variavel sao dividios pelo numero total de obserções do dataframe original e multiplicados por 100

#Ordena os valores com base na variabel 'Valores_Faltantes' em ordem decrescente
df_isna.sort_values(by='Valores_Faltantes', ascending = False)

#A variavel valores faltantes relativos indica em porcentagem a quantidade de dados nulos (ou NaN)
#Assim fica bem visível que no dataframe original, temos variaveis com 98% dos dados valtantes,
#Assim, tenho por mim que variaveis com tantos valores nulo nao podem ser analisados
```

```python
#Com base nos meus próprios critério, irei remover do dataframe original, todas as variveis que possuem valores faltantes maior do que 50% 
#Com base no dataframe acima, eu poderia de dropar manualmente derrubar as 4 primeiras variaveis
#Mais seria interessante criar uma lógica que fizesse isso, pois poderia ser um banco de dados com milhares de variaveis

#
print(f'Numero de colunas antes do drop {df.shape[1]}\n')

for i in df.columns:
    if (len(df[i][df[i].isna() == True]) / df.shape[0]) * 100 > 50.0:
        del df[i]

print(f'Numero de colunas depois do drop {df.shape[1]}')


#Conforme espera, 4 variaveis foram removidas do dataframe
```

```python
#Test para substituição de observações com valores NaN por qualquer outro valor
#Neste caso, utilizarei a variavel onde armazenam valores de email
#Conforme observações anterior, a variavel apresenta uma quantidade 35% de valores NAn

#Selectionando e substituindo todos os falores NaN por 'noemail@noemail.com'
df['EMAIL_RESP'][df['EMAIL_RESP'].isna() == True] = 'noemail@noemail.com'
df
```

```python
#Cruzando dados
np.round(
    pd.crosstab(df['MOTIVO_CANCEL'], 
                df['SIT'], #Selecionando as variaveis a serem cruzadas
                normalize=True).sort_values(by='CANCELADA',  #Ordenando pela coluna 'CANCELADA' em ordem decrescente
                                            ascending=False) * 100 
)
```
